Implementation of a program with quadratic objective

$$
\max z = -x^2 + 2 x  \\
s.t \\
x + y <= 6
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(2,7,100)
plt.plot(x, - (x - 4) ** 2 +  x, '-r')
plt.plot(x,  - x + 6, '-r')
plt.xlabel('x', color='#1C2833')
plt.ylabel('z', color='#1C2833')
plt.grid()
plt.show()
```

Lagrangian
$$
\mathcal{L}(\lambda) = \max_{x, y} -x^2 + 2x + \lambda * (6 - x - y)
$$
And its dual function :
$$
min_{\lambda} \mathcal{L}(\lambda) = \min_{\lambda} \max_{x, y} -x^2 + 2x + \lambda * (6 - x - y)
$$
